```python
def split_text_into_chunks(text, chunk_size):
    chunks = []
    current_pos = 0
    text_length = len(text)

    while current_pos < text_length:
        next_pos = current_pos + chunk_size
        if next_pos < text_length and text[next_pos] != '. ':
            # Move back to the last space character
            while next_pos > current_pos and text[next_pos] not in {'. ', '\n'}:
                next_pos -= 1
        if next_pos == current_pos:
            # In case there are no spaces in the chunk, force split
            next_pos = current_pos + chunk_size
        chunks.append(text[current_pos:next_pos].strip())
        current_pos = next_pos
    
    return chunks
```

```python
import random

def generate_sentences(num_sentences):
    subjects = [
        "The cat", "A dog", "The bird", "A man", "The woman", "The car", "A child", 
        "The teacher", "The student", "A programmer", "An artist", "The musician", 
        "A scientist", "The engineer", "A doctor", "The nurse", "A driver", "The chef", 
        "An athlete", "The dancer"
    ]
    verbs = [
        "jumps", "runs", "flies", "eats", "sleeps", "drives", "writes", "reads", 
        "plays", "watches", "creates", "builds", "fixes", "teaches", "learns", "cooks", 
        "dances", "sings", "paints", "designs"
    ]
    objects = [
        "over the fence", "in the park", "through the window", "on the roof", "under the tree", 
        "in the garden", "at the beach", "in the office", "at home", "in the classroom", 
        "in the library", "at the hospital", "on the street", "in the city", "in the village", 
        "on the mountain", "by the river", "near the lake", "under the bridge", "at the store"
    ]
    
    sentences = []
    for _ in range(num_sentences):
        subject = random.choice(subjects)
        verb = random.choice(verbs)
        object = random.choice(objects)
        sentence = f"{subject} {verb} {object}."
        sentences.append(sentence)
    
    return sentences

def write_sentences_to_file(file_path, sentences, label):
    try:
        with open(file_path, 'w', encoding='utf-8') as file:
            for sentence in sentences:
                file.write(f'__label__{label} ' + sentence + '\n')
        print(f"Sentences written to file: {file_path}")
    except Exception as e:
        print(f"Error writing to file: {e}")

# Example usage
num_sentences = 1000  # Number of sentences to generate
file_path = "generated_sentences.txt"  # File to write the sentences to

sentences = generate_sentences(num_sentences)
write_sentences_to_file(file_path, sentences)



```

```python
import spacy
def remove_person_names(text):
    # 加载spaCy的英语模型
    nlp = spacy.load("en_core_web_sm")
    # 使用模型处理文本
    doc = nlp(text)
    # 创建一个新的列表，用于存储没有人名的词
    tokens_without_person_names = []
    
    # 遍历文档中的每个标记
    for token in doc:
        # 如果标记不是人名，则添加到列表中
        if token.ent_type_ != 'PERSON':
            tokens_without_person_names.append(token.text)
    
    # 将列表连接成一个字符串，返回结果
    return " ".join(tokens_without_person_names)

# 示例使用
text = "Libralato, M.; Bellini, A.; Piotto, G.; Nardiello, D.; van der Marel, R.P .; Anderson, J.; Bedin, L.R.; Vesperini, E."
result = remove_person_names(text)
print(result)  # 输出: "and went to the market. and stayed at home."
```
